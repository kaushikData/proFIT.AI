## Extract employee experiences using tags

```python
#for link in r.loc[0:5,'profile']:
experience_df = pd.DataFrame(columns=["profile_url","job_title","company","period","years_exp","experience_description"])

#for i in range(profile_df.shape[0]):
for i in range(2):
    link = profile_df.loc[i,'profile']
    print ("*************************************************************************")
    print (i)
    
    try:

        browser.get(link)
        
        """
        SCROLL_PAUSE_TIME = 5
        #getting scroll height
        last_height = browser.execute_script("return document.body.scrollHeight")
        for i in range(5):
        #Scroll Down to bottom
            browser.execute_script("window.scrollTo(0, document.body.scrollHeight);")

            #Wait to load the page
            time.sleep(SCROLL_PAUSE_TIME)
            # Calculate the new scroll height nd compare last scroll height
            new_height = browser.execute_script("return document.body.scrollHeight")
            if new_height == last_height:
                break
            last_height = new_height 
        """      
            
        src = browser.page_source
        soup = BeautifulSoup(src, 'lxml')
        print (soup)

        exp_section = soup.find('section',class_='pv-profile-section experience-section ember-view')
        print ("exp_section: ", str(exp_section)[:20])
        li_tags = exp_section.find_all('div', class_='display-flex flex-column full-width')
        a_tags = []
        for i in range(len(li_tags)):
            a_tags.append(li_tags[i].find('a'))
        job_titles = list(map(lambda x:x.find('h3').get_text().strip(), a_tags))
        print ("job: ",len(job_titles))
        print (job_titles)
        company = list(map(lambda x:x.find_all('p')[1].get_text().strip(), a_tags))
        print ("company: ",len(job_titles))
        print (company)
        period = list(map(lambda x:x.find_all('h4')[0].find_all('span')[1].get_text().strip(), a_tags))
        print ("period: ",len(period))
        print (period)
        years_exp = list(map(lambda x:x.find_all('h4')[1].find_all('span')[1].get_text().strip(), a_tags))
        print ("period: ",len(years_exp))
        print (years_exp)
        exp_roles_resps = exp_section.find_all("div", class_= "pv-entity__extra-details t-14 t-black--light ember-view")

        experience = []
        for exp_roles_resp in exp_roles_resps:
            for tag in exp_roles_resp.find_all('span'):
                tag.replace_with('')
            experience.append(str(exp_roles_resp).replace("<!-- --></p><!-- --></div>","").split(';">')[1].strip())
        print ("experience: ",len(experience))
        print (experience)

        temp_df = pd.DataFrame({'profile_url':link, 'job_title':job_titles, 'company':company, 'period':period, 'years_exp':years_exp, 'experience_description':experience[:len(job_titles)]})
        experience_df = pd.concat([experience_df, temp_df])
        experience_df.to_csv("../data/experience_data.csv")
    except Exception as e:
        print (e)

    
    
```
